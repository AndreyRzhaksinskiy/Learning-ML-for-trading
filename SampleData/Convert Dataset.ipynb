```python
import pandas as pd
import numpy as np
```

```python
%%time
# Считать собранный датасет
data = pd.read_csv('..\..\T21\df_consolidated.csv')
data['date'] = pd.to_datetime(data['date'], format='%Y%m%d %H:%M:%S')
data = data.set_index(['date'])
data.sort_index()
print(data.shape)
print('Tickers ',len(data.ticker.unique()))
```

```python
data.head()
```

```python
df = data.sample(1000000).sort_index()  #(12324790, 197)
df = df[['ticker','close']]
df = df.reset_index()
df = df.pivot(index='date', columns='ticker', values='close')
df = df.fillna(method = 'ffill')
df = df.fillna(method = 'bfill') ## temporary approach
print(df.shape)
print(df.isnull().sum())
print(df.isnull().sum().sum())
print( 'Inf -> ', df.columns.to_series()[np.isinf(df).any()] ) # Check Inf
df = df.reset_index()
```

```python
compression_opts = dict(method='zip', archive_name='..\SampleData\stock_5min_v20220613.csv')  
df.to_csv('..\SampleData\stock_5min_v20220613.zip', compression=compression_opts)
```
